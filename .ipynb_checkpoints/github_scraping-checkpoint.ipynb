```python
import requests
import time

access_token = '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'
base_url = 'https://api.github.com/search/repositories'
results_per_page = 100

# Define the date range for the desired 12 months of 2020
start_year = 2020
start_month = 1
end_year = 2020
end_month = 12

repositories = []

# Iterate over the 12 months of 2020
for month in range(start_month, end_month + 1):
    # Calculate the start and end dates for each month
    start_date = f'{start_year}-{month:02d}-01'
    end_date = f'{start_year}-{month:02d}-31'

    # Create the API URL with the date range and results per page
    query = f'created:{start_date}..{end_date}'
    url = f'{base_url}?q={query}&sort=stars&order=desc&per_page={results_per_page}'

    # Make the initial API request
    response = requests.get(url, params={'Authorization': f'Bearer {access_token}'})

    # Check the response status code
    if response.status_code == 200:
        data = response.json()
        items = data.get('items', [])

        # Extract repositories from the initial response
        for item in items:
            repositories.append( {
                    'name': item['name'],
                    'url': item['html_url'],
                    'description': item['description'],
                    'stars': item['stargazers_count'],
                    'created_at': item['created_at'],
                    'language': item.get('language', ''),
                    'forks': item['forks'],
                    'watchers': item['watchers'],
                    'open_issues': item['open_issues']}
            )                

        # Fetch remaining repositories using pagination
        total_count = data.get('total_count', 0)
        total_pages = min(total_count // results_per_page, 10)  # Maximum 10 pages (1,000 repositories)
        for page in range(2, total_pages + 1):
            # Create the API URL for the next page
            page_url = f'{url}&page={page}'

            # Make the API request for the next page
            response = requests.get(page_url, params={'Authorization': f'Bearer {access_token}'})

            # Check the response status code
            if response.status_code == 200:
                data = response.json()
                items = data.get('items', [])

                # Extract repositories from the current page
                for item in items:
                    repositories.append(item['name'])
            else:
                # Request failed
                print(f'Request failed with status code: {response.status_code}')
                break

            # Introduce a delay between requests to respect rate limits
            
    # Check if the desired 12,000 repositories have been reached
    print("pls wait am taking a nap...ZzZz")
     
    time.sleep(30)  # Adjust the sleep duration as needed

    if len(repositories) >= 2000:
        print('its a break')
        break

# Print the fetched repositories
# for repo in repositories:
#     print(repo)

print('fin..')
```
